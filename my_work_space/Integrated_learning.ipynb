```python
import pickle
import numpy as np
```

```python
# 两个模型分别得到的pkl文件路径
p13 = '/home/fate/openmmlab/mmclassification_0/my_work_space/experiment13/result.pkl'
p19 = '/home/fate/openmmlab/mmclassification_0/my_work_space/experiment19/result.pkl'
```

```python
res = {}
labels = ['nv', 'mel']
pred_class = []
with open(p13, 'rb') as f:
    content13 = pickle.load(f)
    class_scores13 = content13['class_scores']

with open(p19, 'rb') as f:
    content19 = pickle.load(f)
    class_scores19 = content19['class_scores']

class_scores = (class_scores13 + class_scores19) / 2
res['class_scores'] = class_scores

pred_label = np.argmax(class_scores , axis = 1)
pred_label

pred_score = np.max(class_scores, axis=1)
pred_score

for x in pred_label:
    pred_class.append(labels[x])
pred_class
res['class_scores'] = class_scores
res['pred_score'] = pred_score
res['pred_label'] = pred_label
res['pred_class'] = pred_class
with open('experiment44/avg_result.pkl', 'wb') as f:
    pickle.dump(res, f)

```
