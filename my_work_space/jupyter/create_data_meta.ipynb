创建元数据信息，供mmcls训练使用

```python
import os 
import pandas as pd
import json
```

```python
# csv 文件
meta_csv = 'F:/work_spaces/vscode/melanoma/datasets/ISIC2018/ISIC2018_Task3_Training_GroundTruth/ISIC2018_Task3_Training_GroundTruth.csv'
# 图像所在文件夹
# img_pth_prefix = 'F:/work_spaces/vscode/melanoma/datasets/ISIC2018/ISIC2018_Task3_Training_Input'
# 转换后的json格式
imgs_metainfo = {
    'metainfo':
        {
            'classes': ('mel', 'nv'), # 'nv' 的类别序号为 0，'mel' 为 1。
            
        },
    'data_list': [ ]
}

save_file = './meta.json'

```

```python
df = pd.read_csv(meta_csv)
df = df.loc[(df['NV'] == 1) | (df['MEL'] == 1)]
df
```

```python
data_list = []
for index, row in df.iterrows():
    data = {
        'img_path': "",
        'img_label': 0,
    }
    img_name = row['image']
    if row['MEL'] == 1:
        clazz = 1
        pass
    elif row['NV'] == 1:
        clazz = 0
        pass
    data['img_path'] = f'{img_name}.jpg'
    data['img_label'] = clazz
    data_list.append(data)

imgs_metainfo['data_list'] = data_list
with open(save_file, 'w') as f:
    json.dump(imgs_metainfo, f)
```
